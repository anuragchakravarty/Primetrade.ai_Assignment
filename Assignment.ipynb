## Data Cleaning and Dataset Integration

```python
trader_df['Timestamp IST'] = pd.to_datetime(
    trader_df['Timestamp IST'],
    dayfirst=True
)

# Create date column
trader_df['date'] = trader_df['Timestamp IST'].dt.normalize()

# Convert sentiment date
sentiment_df['date'] = pd.to_datetime(sentiment_df['date'])

# Merge
merged_df = pd.merge(
    trader_df,
    sentiment_df[['date', 'classification']],
    on='date',
    how='left'
)

print("Merged Shape:", merged_df.shape)
print("Missing Sentiment:", merged_df['classification'].isnull().sum())
```

##### Here above (Missing Sentiment: 6). This is perfectly acceptable. Those dates likely don't exist in the sentiment dataset

## Basic Analysis

```python
# Total trades
print("Total Trades:", len(merged_df))
```

```python
# Sentiment distribution
print("Trades by Market Sentiment:")
print(merged_df['classification'].value_counts())
```

```python
# Average Closed PnL by sentiment
print("Average Closed PnL by Sentiment:")
print(merged_df.groupby('classification')['Closed PnL'].mean().sort_values(ascending=False))
```

```python
# Average leverage/position (if leverage column isn't available)
print("Average Trade Size (USD):")
print(merged_df.groupby('classification')['Size USD'].mean())
```

```python
# Buy vs Sell trades
print("Buy/Sell Distribution:")
print(merged_df['Side'].value_counts())
```

```python
# Most traded coins
print("Top 10 Most Traded Coins:")
print(merged_df['Coin'].value_counts().head(10))
```

## Visualization

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=merged_df,
    x='classification',
    order=merged_df['classification'].value_counts().index
)

plt.title("Distribution of Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Number of Trades")
plt.xticks(rotation=20)
plt.show()
```

```python
plt.figure(figsize=(8,5))

avg_pnl = merged_df.groupby('classification')['Closed PnL'].mean().sort_values()

sns.barplot(
    x=avg_pnl.index,
    y=avg_pnl.values
)

plt.title("Average Closed PnL by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Average Closed PnL")
plt.xticks(rotation=20)
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    data=merged_df,
    x='classification',
    y='Closed PnL'
)

plt.title("Closed PnL Distribution Across Market Sentiments")
plt.xlabel("Market Sentiment")
plt.ylabel("Closed PnL")
plt.xticks(rotation=20)

plt.show()
```

```python
plt.figure(figsize=(6,6))

merged_df['Side'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90
)

plt.ylabel("")
plt.title("Buy vs Sell Trades")

plt.show()
```

```python
top_traders = (
    merged_df
    .groupby('Account')['Closed PnL']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_traders.values,
    y=top_traders.index
)

plt.title("Top 10 Most Profitable Traders")
plt.xlabel("Total Closed PnL")
plt.ylabel("Trader Account")

plt.show()
```
